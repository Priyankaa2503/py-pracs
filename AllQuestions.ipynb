```python
s = "Hello! How are you? 123" 

num_digits = 0
num_special = 0
num_vowels = 0
num_consonants = 0

vowels = set("aeiouAEIOU")

for char in s:
    if char.isdigit():
        num_digits += 1
    elif not char.isalpha() and not char.isspace():
        num_special += 1
    elif char in vowels:
        num_vowels += 1
    else:
        num_consonants += 1
print("Number of digits:", num_digits)
print("Number of special symbols:", num_special)
print("Number of vowels:", num_vowels)
print("Number of consonants:", num_consonants)
```

Q37

```python
def find_pythagorean_triplets(range_limit):
    triplets = []
    for a in range(1, range_limit+1):
        for b in range(a+1, range_limit+1):
            c = (a**2 + b**2) ** 0.5
            if c.is_integer() and c <= range_limit:
                triplets.append((a, b, int(c)))
    return triplets

range_limit = 20
triplets = find_pythagorean_triplets(range_limit)
for triplet in triplets:
    print(triplet)
```

Q39

```python
def minimum_moves_to_sort(A):
    def longest_increasing_subsequence(A):
        n = len(A)
        dp = [1] * n
        for i in range(1, n):
            for j in range(i):
                if A[i] > A[j]:
                    dp[i] = max(dp[i], dp[j] + 1)
        return max(dp)
    n = len(A)
    mis_length = longest_increasing_subsequence(A)
    min_moves = n - mis_length
    return min_moves

A = [4, 5, 2, 1, 3]
print("Original array:", A)
moves = minimum_moves_to_sort(A)
print("Minimum number of moves required:", moves)
```

```python
number = int(input("Enter your number: "))
print("Number : ",number)
arr = []
for i in range(number):
    arr.insert(i,(int(input("Enter element: "))))
print("Array : ",arr)
count = 0 
for i in range(number-1):
    j=i+1
    while(j<number):
        if(arr[j]<arr[i]):
            arr[j],arr[i]=arr[i],arr[j]
            count+=1
        j=j+1
print("Sorted Array : ",arr)
print(count)
```
